```python
import pandas as pd

# Load the dataset
df = pd.read_csv("College.csv")

# Display basic information about the dataset
print(df.info())

# Summary statistics
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.histplot(df["Top10perc"], bins=20, kde=True)
plt.title("Distribution of Top10perc")
plt.xlabel("Top10perc")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="Private", y="Accept", data=df)
plt.title("Acceptance Rate by Private/Public")
plt.xlabel("Private")
plt.ylabel("Acceptance Rate")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already loaded your DataFrame 'df'

# Create a pairplot for the specified columns
sns.pairplot(df[["Top10perc", "Outstate", "Grad.Rate"]])

# Set the title for the pairplot
plt.title("Pairplot of Top10perc, Outstate, and Grad.Rate")

# Adjust the figure layout
plt.tight_layout()

# Show the plot
plt.show()
```
